```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import io
from matplotlib import pyplot as plt
```

```python
directoryOne = 'datasets/001-Youth_Not_in_Education_and_Not_in_Employment_15-24_years old_percent_of_Youth_Household_Population_by_Sex.csv'
```

```python
datasetOne= pd.read_csv(directoryOne)
datasetOne = datasetOne.rename(columns={'Column1': ' ', 'Column2': ' '})
datasetOne

```

Data Visualization

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))


df_line = datasetOne.iloc[0, 1:].astype(int)  
df_line.plot(kind='line', ax=ax1, marker='o', linestyle='-', color='green')
ax1.set_ylabel('Percent Share of Youth')
ax1.set_title('Percent Share of Youth not in Education and not in Employment')


df_bar = datasetOne.iloc[1, 1:].astype(int)  
df_bar.plot(kind='bar', ax=ax2, width=0.8, color='blue', alpha=0.7)
ax2.set_ylabel('Number of Youth (in thousands)')
ax2.set_title('Total Youth Not in Education and Not in Employment')

plt.tight_layout()
plt.show()
```

Datasets of GDP Growth by Industry 2018-2022

```python
directoryTwo = 'datasets/002 Gross Domestic Product, by Industry, Growth Rates.csv'
```

```python
datasetTwo= pd.read_csv(directoryTwo, skiprows=1)
datasetTwo.head(2000)
```

Data Cleaning
Deleting of Unnecessary rows [and metadata]

```python
datasetTwo = datasetTwo.iloc[2:22]
datasetTwo = datasetTwo.dropna(axis=0, how="all")
datasetTwo = datasetTwo.rename(columns={'Unnamed: 0': 'Industry', 'Unnamed: 1': '2010-2011',  'Unnamed: 2' : '2011-2012', 'Unnamed: 3' : '2012-2013', 'Unnamed: 4': '2013-2014', 'Unnamed: 5' : '2014-2015', 'Unnamed: 6': '2015-2016',
'Unnamed: 7': '2016-2017', 'Unnamed: 8': '2017-2018', 'Unnamed: 9' : '2018-2019', 'Unnamed: 10' : '2019-2020', 'Unnamed: 11' : '2020-2021', 'Unnamed: 12': '2021-2022'})

datasetTwo
```

Data Visualization

```python
datasetTwo.set_index('Industry', inplace=True)


dataset = datasetTwo.transpose()

dataset = dataset.apply(pd.to_numeric)

dataset = dataset[dataset.columns.sort_values()]

plt.figure(figsize=(12, 6))
for column in dataset.columns:
    plt.plot(dataset.index, dataset[column], marker='o', label=column)

plt.title('Growth Rates of Industries over Time')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=0)
plt.grid(True)

y_min = dataset.min().min()  
y_max = dataset.max().max()  
plt.ylim(y_min - 2, y_max + 2)  


plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=4)

plt.show()
```

Almost every dataset has the same behavioral pattern as the other. The only economic factor that showed similarity with the inflation rate and consumer price index was the exportation of goods dataset. Based on this, there is an observe correlation between the decrease of exportation rate on primary commodities and the overall inflation rate. Thus, the researchers recommend measures in relation to the trade of the country. The first recommendation is to support the production of export products. To do this, the government should focus on the development of MSMEs since it compose 99.58% of all businesses in the country. As a result, domestic trade will be improved which would lead to a better control on the value of the Philippine Peso.
